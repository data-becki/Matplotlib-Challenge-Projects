```python
# Create a function to plot rankings over time for 1 roller coaster
def rankings_over_time(coaster_name, park_name, ranking_df):
    rank = ranking_df[(ranking_df.Name == coaster_name) & (ranking_df.Park == park_name)]
    plt.plot(rank['Year of Rank'], rank['Rank'])
    plt.title('Rankings for ' + coaster_name + ' coaster at ' + park_name)
    plt.gca().invert_yaxis()
    plt.xlabel('Year')
    plt.ylabel('Rank')
    plt.show()

# Create a plot of El Toro ranking over time
rankings_over_time('El Toro', 'Six Flags Great Adventure', wood_coaster_rank)
```

```python
# Create a function to plot rankings over time for 2 roller coasters
def two_rankings(coaster_1, coaster_2, park_1, park_2, ranking_df):
    rank_1 = ranking_df[(ranking_df.Name == coaster_1) & (ranking_df.Park == park_1)]
    rank_2 = ranking_df[(ranking_df.Name == coaster_2) & (ranking_df.Park == park_2)]
    plt.plot(rank_1['Year of Rank'], rank_1['Rank'], label = coaster_1)
    plt.plot(rank_2['Year of Rank'], rank_2['Rank'], label = coaster_2)
    plt.gca().invert_yaxis()
    plt.title('Rankings for ' + coaster_1 + ' & ' + coaster_2 + ' coasters')
    plt.xlabel('Year')
    plt.ylabel('Rank')
    plt.legend()
    plt.show()

# Create a plot of El Toro and Boulder Dash roller coasters
two_rankings('El Toro', 'Boulder Dash', 'Six Flags Great Adventure', 'Lake Compounce', wood_coaster_rank)
```

4. Write a function that will plot the ranking of the top `n` ranked roller coasters over time as lines. Your function should take a number `n` and a ranking DataFrame as arguments. Make sure to include informative labels that describe your visualization.

   For example, if `n == 5`, your function should plot a line for each roller coaster that has a rank of `5` or lower.
   
   Call your function with a value of `n` and either the wood ranking or steel ranking DataFrame.

```python
# Create a function to plot top n rankings over time
def top_n_ranked(n, ranking_df):
    top_ranked = ranking_df[(ranking_df.Rank <= n)]
    plt.subplots(figsize=(10,5))
    for coaster in set(top_ranked['Name']):
        coaster_rankings = top_ranked[top_ranked['Name'] == coaster]
        plt.plot(coaster_rankings['Year of Rank'], coaster_rankings['Rank'], label=coaster)
    plt.gca().invert_yaxis()
    plt.title('Top ' + str(n) + ' rankings over time')
    plt.xlabel('Year')
    plt.ylabel('Rank')
    plt.legend()
    plt.show()

# Create a plot of top n rankings over time
top_n_ranked(3, wood_coaster_rank)
```

```python
# Create a function to plot histogram of column values
def hist_values(column_name, df):
    plt.hist(df[column_name].dropna())
    plt.xlabel(column_name)
    plt.ylabel('Count')
    plt.title('Histogram of ' + column_name)
    plt.show()

# Create histogram of roller coaster speed
hist_values('speed', coaster_data)

# Create histogram of roller coaster length
hist_values('length', coaster_data)

# Create histogram of roller coaster number of inversions
hist_values('num_inversions', coaster_data)

# Create a function to plot histogram of height values
def plot_height_histogram(coaster_df):
    heights = coaster_df[coaster_df['height'] <= 140]['height'].dropna()
    plt.hist(heights)
    plt.title('Histogram of Roller Coaster Height')
    plt.xlabel('Height')
    plt.ylabel('Count')
    plt.show()
    plt.close()

# Create a histogram of roller coaster height
plot_height_histogram(coaster_data)
```

7. Write a function that creates a bar chart showing the number of inversions for each roller coaster at an amusement park. Your function should take the roller coaster DataFrame and an amusement park name as arguments. Make sure to include informative labels that describe your visualization.

   Call your function with the roller coaster DataFrame and amusement park name.

```python
# Create a function to plot inversions by coaster at park
def inversion(coaster_df, park_name):
    # select rows of df that belong to desired park
    park_coasters = coaster_df[coaster_df['park'] == park_name]
    
    # sorted df by # inversion in desc order
    park_coasters = park_coasters.sort_values('num_inversions', ascending=False)
    
    # get names and numbers to create bar chart
    coaster_names = park_coasters['name']
    number_inversions = park_coasters['num_inversions']
    
    plt.bar(coaster_names, number_inversions)
    plt.title('Number of Inversions in Roller Coasters at ' + park_name)
    plt.xlabel('Coaster Names')
    plt.ylabel('Number of Inversions')
    plt.xticks(rotation=65)
    plt.show()

# Create barplot of inversions by roller coasters
inversion(coaster_data, 'Holiday World')
inversion(coaster_data, 'Parc Asterix')
```

8. Write a function that creates a pie chart that compares the number of operating roller coasters (`'status.operating'`) to the number of closed roller coasters (`'status.closed.definitely'`). Your function should take the roller coaster DataFrame as an argument. Make sure to include informative labels that describe your visualization.

   Call your function with the roller coaster DataFrame.

```python
# Create a function to plot a pie chart of status.operating
def operational(df):
    operating_coasters = df[df['status'] == 'status.operating']
    closed_coasters = df[df['status'] == 'status.closed.definitely']
    num_operating_coasters = len(operating_coasters)
    num_closed_coasters = len(closed_coasters)
    status_counts = [num_operating_coasters,num_closed_coasters]
    plt.pie(status_counts, autopct='%0.1f%%', labels=['Operating', 'Closed'], explode=(0, 0.1), shadow=True)
    plt.show()
    
# Create pie chart of roller coasters
operational(coaster_data)
```

9. `.scatter()` is another useful function in matplotlib that you might not have seen before. `.scatter()` produces a scatter plot, which is similar to `.plot()` in that it plots points on a figure. `.scatter()`, however, does not connect the points with a line. This allows you to analyze the relationship between two variables. Find [`.scatter()`'s documentation here](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html).

   Write a function that creates a scatter plot of two numeric columns of the roller coaster DataFrame. Your function should take the roller coaster DataFrame and two-column names as arguments. Make sure to include informative labels that describe your visualization.
   
   Call your function with the roller coaster DataFrame and two-column names.

```python
# Create a function to plot scatter of any two columns
def scatter_num(df, column_1, column_2):
    plt.scatter(df[column_1], df[column_2])
    plt.title('Scatter Plot of ' + column_1 + ' & ' + column_2)
    plt.xlabel(column_1)
    plt.ylabel(column_2)
    plt.show()
    plt.close()

scatter_num(coaster_data, 'num_inversions', 'length')
    
# Create a function to plot scatter of speed vs height
def speed_height(df):
    df = df[df['height'] < 140]
    plt.scatter(df['height'], df['speed'])
    plt.title('Scatter Plot of Speed vs. Height')
    plt.xlabel('height')
    plt.ylabel('speed')
    plt.show()


# Create a scatter plot of roller coaster height by speed
speed_height(coaster_data)
```

```python
# Do different seating types result in higher/faster/longer roller coasters? 
def seat_speed(df):
    df = df[df['speed'] > 0]
    plt.scatter(df['seating_type'], df['speed'], alpha = 0.4)
    plt.title('Scatter Plot of Speed vs. Seating Type')
    plt.xlabel('Seating Type')
    plt.ylabel('Speed')
    plt.xticks(rotation=65)
    plt.show()

def height_speed(df):
#     df = df[df['height'] >= 0]
    plt.scatter(df['seating_type'], df['height'], alpha = 0.2)
    plt.title('Scatter Plot of Height vs. Seating Type')
    plt.xlabel('Seating Type')
    plt.ylabel('Height')
    plt.xticks(rotation=65)
    plt.show()
    
def length_speed(df):
    df = df[df['length'] > 0]
    plt.scatter(df['seating_type'], df['length'], alpha = 0.4)
    plt.title('Scatter Plot of Length vs. Seating Type')
    plt.xlabel('Seating Type')
    plt.ylabel('Length')
    plt.xticks(rotation=65)
    plt.show()
    
seat_speed(coaster_data)
height_speed(coaster_data)
length_speed(coaster_data)
```

*Length possibly has an effect on seating type, both 'Sit Down' and 'Alpine' have the longer lengths and they are both seated positions. Speed and Height have an effect on seating type - 'Sit Down' has the wider range of values.*

```python
def seat_comp(df, column_name):
    # sorted df by values in desc order
    df = df.sort_values(column_name, ascending=False)
    y = df[column_name]
    x = df['seating_type']
    
    plt.bar(x, y)
    plt.title(column_name + ' vs. Seating Type')
    plt.xlabel('Seating Type')
    plt.ylabel(column_name)
    plt.xticks(rotation=65)
    plt.show()

seat_comp(coaster_data, 'speed')
seat_comp(coaster_data, 'height')
seat_comp(coaster_data, 'length')
```

```python
park_length = []
for name in park_10_coasters['park']:
    length = coaster_data[coaster_data['park'] == name].length.mean()
    park_length.append(length)

df_length = pd.DataFrame(park_length, park_name).reset_index()
df_length.columns = ['park', 'avg_length']
df_length.sort_values(by = 'avg_length', ascending = False, inplace = True)

plt.figure(figsize=(10, 5))
plt.bar(df_length['park'], df_length['avg_length'])
plt.xticks(rotation = 85)
plt.title('Parks\' average length of roller coasters')
plt.ylabel('Length in m')
plt.xlabel('Park')
plt.show()
```

```python
park_inv = []
for name in park_10_coasters['park']:
    inv = coaster_data[coaster_data['park'] == name].num_inversions.mean()
    park_inv.append(inv)

df_inv = pd.DataFrame(park_inv, park_name).reset_index()
df_inv.columns = ['park', 'avg_inv']
df_inv.sort_values(by = 'avg_inv', ascending = False, inplace = True)

plt.figure(figsize=(10, 5))
plt.bar(df_inv['park'], df_inv['avg_inv'])
plt.xticks(rotation = 85)
plt.title('Parks\' average number of inversions in their roller coasters')
plt.ylabel('Number of inversions')
plt.xlabel('Park')
plt.show()
```
